```r
library(tidyverse)
library(readxl)
library(stringr)

month_th <- c("มกราคม",
              "กุมภาพันธ์",
              "มีนาคม",
              "เมษายน",
              "พฤษภาคม",
              "มิถุนายน",
              "กรกฎาคม",
              "สิงหาคม",
              "กันยายน",
              "ตุลาคม",
              "พฤศจิกายน",
              "ธันวาคม")

get_headers <- function(path, ...){
    dat <- read_excel(path, ...)
    header <- t(apply(dat, 1, na.locf, na.rm = FALSE))
    return(paste(header[1,], header[2,], sep='_'))
}

read_price_w <- function(path, ...){
    first_line <- read_excel(path, range="A1", col_names=FALSE)
    year_ <- str_sub(first_line, -4)
    month_ <- str_extract(first_line, paste(month_th, collapse="|"))

    col_names <- get_headers(path, col_names=FALSE, skip=2, n_max=2) %>%
        gsub("_NA", "", .) %>%
        gsub("สัปดาห์ที่", "", .)

    df <- read_excel(path, col_names=col_names, skip=4, na = c("-","",0)) %>%
        gather("region_week", "amount", -1) %>%
        drop_na() %>%
        separate(region_week, sep="_", into=c("region","week")) %>%
        rename(price_name = รายการ) %>%
        arrange(price_name) %>%
        mutate(year = year_) %>%
        mutate(month_th = month_)
    return(df)
}
```

```r
root <- "../data/R3_prices/2.ราคารายสัปดาห์/ราคาพืชเศรษฐกิจรายสัปดาห์"
files <- list.files(root)
```

```r
library(ddpcr)
```

```r
data <- data.frame()
for (i in seq_along(files)){
    path <- paste(root, files[i], sep="/")
    dt <- read_price_w(path)
    data <- rbind(data, dt)
}
```

```r
unique(data$price_name)
```

```r
data %>%
    filter(price_name == "ข้าวเปลือกเจ้าหอมมะลิ") %>%
    group_by(year, month_th, week) %>%
    summarize(amount = mean(amount))
```
